```python
import pickle as pkl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_pickle_file(file):
    pickle_data = pd.read_pickle(file)
    return pickle_data
```

```python
cs_vw = read_pickle_file('cs_vw.pkl')
cs_ew = read_pickle_file('cs_ew.pkl')
```

# Print Sharpe Ratios

## vw

```python
uni_sr = [] 
for c in cs_vw.char_list+['me']:
    expr = "cs_vw.port_%s.uni_ret.to_csv('./data_Jan2021/chars60/sorted_portfolio_vw/unisort/unisort_vw_%s.csv')"%(c,c)
    exec(expr)
    expr = "uni_sr.append(cs_vw.port_%s.uni_ls_sr)"%(c)
    exec(expr)

bi_sr = []
for c in cs_vw.char_list:
    expr = "cs_vw.port_%s.bi_ret.to_csv('./data_Jan2021/chars60/sorted_portfolio_vw/bisort/bisort_vw_%s.csv')"%(c,c)
    exec(expr)
    expr = "bi_sr.append(cs_vw.port_%s.bi_ls_sr)"%(c)
    exec(expr)

uni_sr = pd.Series(uni_sr, index = cs_vw.char_list+['me'])
bi_sr = pd.Series(bi_sr, index = cs_vw.char_list)
```

```python
plt.figure(figsize=(10,5))
plt.bar(uni_sr.index, uni_sr.values)
plt.title('61 Unisort Value-weight Return Annualized Sharpe Ratios')
plt.xticks(rotation='vertical')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(10,5))
plt.bar(bi_sr.index, bi_sr.values)
plt.title('60 Bisort Value-weight Return Annualized Sharpe Ratios')
plt.xticks(rotation='vertical')
plt.show()
plt.close()
```
