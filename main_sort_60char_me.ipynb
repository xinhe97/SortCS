```python
cs_vw.port_me = port(cs_vw.df, 'me', cs_vw.char_list, cs_vw.n, cs_vw.w)
cs_ew.port_me = port(cs_ew.df, 'me', cs_ew.char_list, cs_ew.n, cs_ew.w)

cs_vw.port_me.unisort()
cs_ew.port_me.unisort()

cs_vw.port_me.uni_count = cs_vw.port_me.df.groupby(['date','port_uni_'+cs_vw.port_me.char]).count()['ret'].unstack()
cs_vw.port_me.uni_ret = cs_vw.port_me.value_weight_average('ret',['date','port_uni_'+cs_vw.port_me.char],cs_vw.port_me.w)
cs_vw.port_me.uni_ls = cs_vw.port_me.uni_ret['0'] - cs_vw.port_me.uni_ret[str(cs_vw.port_me.n-1)]
cs_vw.port_me.ls_sign = 'Long Low'
cs_vw.port_me.uni_ls_sr = cs_vw.port_me.uni_ls.mean()/cs_vw.port_me.uni_ls.std()*np.sqrt(len(cs_vw.port_me.uni_ls))

cs_ew.port_me.uni_count = cs_ew.port_me.df.groupby(['date','port_uni_'+cs_ew.port_me.char]).count()['ret'].unstack()
cs_ew.port_me.uni_ret = cs_ew.df.groupby(['date','port_uni_'+cs_ew.port_me.char]).mean()['ret'].unstack()
cs_ew.port_me.uni_ls = cs_ew.port_me.uni_ret['0'] - cs_ew.port_me.uni_ret[str(cs_ew.port_me.n-1)]
cs_ew.port_me.ls_sign = 'Long Low'
cs_ew.port_me.uni_ls_sr = cs_ew.port_me.uni_ls.mean()/cs_ew.port_me.uni_ls.std()*np.sqrt(len(cs_ew.port_me.uni_ls))
```
